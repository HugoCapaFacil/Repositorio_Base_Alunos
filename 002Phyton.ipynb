# Função Print()
a função print() é uma das funções mais importantes e usadas na liguagem python. sua função é, basicamente, exibir mensagens na tela ou enviá-las para outro dispositivo, como imprimir dentro de um arquivo texto.

No Python, Print() é uma função interna, de modo que não é necessario importar nenhuma biblioteca para poder utliza-la. basta chamalá-la e passar os argumentos necessarios.


```python
print(999 + 1) # O python realiza a operação
```

```python
print("20 + 15") # a operação não se realiza
```

```python
print("D'água") # as aspas tem q ser diferentes
```

```python
print("20 + 15 =", 20 + 15)
```

```python
n1 = "1"
n2 = "3"
print(n1 + n2)
```

```python
nome = "Messi"

sobrenome = "Ronaldo"

nome_completo = nome + " "+ sobrenome
print(nome_completo)
```
