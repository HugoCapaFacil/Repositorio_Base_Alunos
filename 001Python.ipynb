#Tutorial Variaveis


## O que são variaveis?


Variaveis são espaços na memoria do computador onde podemos guardar informações (como textos, números, resultados de calculos, etc) para usar depois

Variavel com texto (string)

```python
nome = "Messi"
print(nome)
```

```python
nome = "Cristiano Ronaldo"
```

```python
print(nome)
```

Variavel com numero inteiro (int)

```python
idade = 15
print(idade)
```

Variavel com numero decimal

```python
# Altura:
altura = 1.77 # Lembre-se que sempre use "." não ","
print(altura)
```

```python
pi = 3.14
print(pi)
```

Variavel com valor booleano (bool)

```python
estudante = True
print(estudante)
```

# Nomes de variaveis

Não pode:

* Começar com numero

* usar espaço

* usar acentuação ou caracteres especiais

* usar palavras reversas

## O que é snake case?

snake case é um tipo de escrita usado para nome de variaveis, funções, e nomes em geral do codigo neste estilo as palavras são sempre em mimusculas e a separação é por "_"

## Porque snake_case?

* Deixa o codigo mais facil de ler
* é o padra recomendado pela python
* segue o PEF8, que é o guia de estilo oficial do python



## Exemplos q não usam no snake_case

```python
nomeCompleto = "Maria" # usado em Java
NomeCompleto = "Maria" # estilo PascalCase (Classe de Python)
nomeCompleto = "Maria" # estilo dificil de ler
```
